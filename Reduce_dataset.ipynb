```python
import os
import shutil
import random
```

```python
# Define the source directory (original dataset) and destination directory (new dataset)
source_dir = 'D:/plant-pathology-2021-fgvc8/resized_dataset(256)'
destination_dir = 'D:/plant-pathology-2021-fgvc8/Reduced'
```

```python
# Dictionary to specify the class names and the number of images you want in the new dataset
class_image_counts = {
    'healthy': 1749,
    'scab': 1851,
    'frog_eye_leaf_spot': 1290,
    'rust': 769,
    'complex': 702,
    'powdery_mildew': 584,
    'scab frog_eye_leaf_spot': 386,
    'scab frog_eye_leaf_spot complex': 200,
    'frog_eye_leaf_spot complex': 165,
    'rust frog_eye_leaf_spot': 120,
    'rust complex': 97,
    'powdery_mildew complex': 87
}
```

```python
# Function to create a new dataset
def create_new_dataset(source_dir, destination_dir, class_image_counts):
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)

    for class_name, image_count in class_image_counts.items():
        source_class_dir = os.path.join(source_dir, class_name)
        destination_class_dir = os.path.join(destination_dir, class_name)
        
        if not os.path.exists(destination_class_dir):
            os.makedirs(destination_class_dir)
        
        # Get all images from the source class directory
        images = os.listdir(source_class_dir)
        
        # Shuffle and select the required number of images
        selected_images = random.sample(images, image_count)
        
        # Copy the selected images to the destination folder
        for image in selected_images:
            source_image_path = os.path.join(source_class_dir, image)
            destination_image_path = os.path.join(destination_class_dir, image)
            shutil.copyfile(source_image_path, destination_image_path)

# Call the function to create the new dataset
create_new_dataset(source_dir, destination_dir, class_image_counts)
```
