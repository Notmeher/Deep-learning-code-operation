```python
import os
import shutil
import pandas as pd
```

```python
# Load the CSV file
file_path = 'D:/plant-pathology-2021-fgvc8/train.csv'
data = pd.read_csv(file_path)
```

```python

image_base_dir = 'D:/plant-pathology-2021-fgvc8/train_images'  # Change this to the directory containing the images
output_base_dir = 'D:/plant-pathology-2021-fgvc8/dataset'  # Change this to the desired output directory
```

```python

# Ensure the output base directory exists
os.makedirs(output_base_dir, exist_ok=True)

# Create a folder for each unique label
for label in data.iloc[:, -1].unique():
    label_dir = os.path.join(output_base_dir, str(label))
    os.makedirs(label_dir, exist_ok=True)

# Copy images to corresponding label folders
for index, row in data.iterrows():
    image_name = row['image']  # Change this to the correct column name if different
    label = row.iloc[-1]
    
    # Define source and target paths
    source_path = os.path.join(image_base_dir, image_name)
    target_path = os.path.join(output_base_dir, str(label), image_name)
    
    # Copy the image to the target folder
    if os.path.exists(source_path):
        shutil.copy(source_path, target_path)
    else:
        print(f"Image {image_name} not found in {image_base_dir}")

print("Dataset creation completed.")
```
