```python
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
# Define the source dataset directory
dataset_dir = 'D:/plant-pathology-2021-fgvc8/Reduced'
```

```python

# Define the destination directories for train and validation sets
train_dir = 'D:/plant-pathology-2021-fgvc8/Train'
val_dir = 'D:/plant-pathology-2021-fgvc8/Valid'
```

```python
# Create the destination directories if they don't exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Fetch class names from the source folder
class_names = [d for d in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, d))]

# Iterate over each class and split the images
for class_name in class_names:
    class_dir = os.path.join(dataset_dir, class_name)
    images = os.listdir(class_dir)
    
    # Split into training (80%) and validation (20%)
    train_images, val_images = train_test_split(images, test_size=0.2, random_state=42)
    
    # Create directories for the current class in train and validation folders
    os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(val_dir, class_name), exist_ok=True)
    
    # Copy training images
    for img in train_images:
        shutil.copy(os.path.join(class_dir, img), os.path.join(train_dir, class_name, img))
    
    # Copy validation images
    for img in val_images:
        shutil.copy(os.path.join(class_dir, img), os.path.join(val_dir, class_name, img))

print("Dataset copied into train and validation sets successfully!")
```
